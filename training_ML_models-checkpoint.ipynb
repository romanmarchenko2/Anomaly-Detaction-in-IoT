### 1.1. Підготовка та аналіз даних для мультикласової класифікації

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import auc, average_precision_score, classification_report, precision_recall_curve, roc_curve
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
from tqdm.auto import tqdm

TRAIN_PATH = (
    r'C:\Users\pumac\Desktop\Diplom_Code\ML-IOT-malware-analysis\datasets\my_train.parquet'
)
TEST_PATH = (
    r'C:\Users\pumac\Desktop\Diplom_Code\ML-IOT-malware-analysis\datasets\my_test.parquet'
)
train = pd.read_parquet(TRAIN_PATH, engine="fastparquet")
test = pd.read_parquet(TEST_PATH, engine="fastparquet")
X_train = train.loc[:, train.columns != "label"]
y_train = train["label"]
X_test = test.loc[:, test.columns != "label"]
y_test = test["label"]

labels = [0, 1, 2, 5, 7, 8, 9]
label_dict = {l: i for i, l in enumerate(labels)}
```

### 1.2. Cтворення матриці розсіювань між усіма парами ознак в датасеті

```python
sns.pairplot(train, hue='label')
plt.show()
```

### 2.1. Тренування моделі випадкового лісу на тренувальному наборі даних та оцінка її точність на тренувальному та тестовому наборах

```python
from sklearn.ensemble import RandomForestClassifier

RANDOM_STATE=42

rndforest = RandomForestClassifier(random_state=RANDOM_STATE)
rndforest.fit(X_train, y_train)

train_score = rndforest.score(X_train, y_train)
test_score = rndforest.score(X_test, y_test)

print(f'Accuracy on training set: {train_score:.2%}')
print(f'Accuracy on test set: {test_score:.2%}')
```

### 2.2. Використання натренованої моделі випадкового лісу для прогнозування класів на тестовому наборі даних (X_test). Отримано звіт про класифікацію

```python
y_pred = rndforest.predict(X_test)
print(classification_report(y_test, y_pred, digits=4))
```

### 2.3. Використання GridSearchCV для пошуку оптимальних параметрів для моделі класифікатора RandomForestClassifier

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [100, 200, 500],
    'max_features': ['sqrt', 0.5, None],
    'criterion' :['gini', 'entropy']
}
model =  RandomForestClassifier(random_state=RANDOM_STATE)
CV_rfc = GridSearchCV(estimator=model, param_grid=param_grid)
CV_rfc.fit(X_train, y_train)
```

### 2.4. Отримані найкращі гіперпараметри для моделі випадкового лісу 

```python
CV_rfc.best_params_
```

### 2.5. Тренування оптимізаваної моделі випадкового лісу за допомогою GridSearchCV на тренувальному наборі даних та оцінка її точність на тренувальному та тестовому наборах

```python
y_pred = CV_rfc.predict(X_test)
print(classification_report(y_test, y_pred, digits=4))
```

### 2.6. Тренування моделі випадкового лісу з оптимальними параметрами для класифікації бінаризованих міток, візуалізація криві точності-повноти та ROC для кожного класу, і зберігання графіків у файл SVG

```python
y_train_bin = label_binarize(y_train, classes=labels)
y_test_bin = label_binarize(y_test, classes=labels)

clf = OneVsRestClassifier(RandomForestClassifier(random_state=RANDOM_STATE, **CV_rfc.best_params_))

clf.fit(X_train, y_train_bin)
y_score = clf.predict_proba(X_test)
fig, axes = plt.subplots(
    nrows=1,
    ncols=2,
    figsize=(12, 6),
    layout="constrained",
)

ap = average_precision_score(y_test_bin, y_score, average=None)
for i in labels:
    precision_i, recall_i, _ = precision_recall_curve(
        y_test_bin[:, label_dict[i]],
        y_score[:, label_dict[i]]
    )
    axes[0].plot(recall_i, precision_i, lw=2, label=f'Class {i} (AP = {ap[label_dict[i]]:.7f})')
    
axes[0].set_xlabel("Recall")
axes[0].set_ylabel("Precision")
axes[0].legend(loc="Best")
axes[0].set_title("Precision vs Recall curve (OvR)")

for i in labels:
    fpr_i, tpr_i, _ = roc_curve(
        y_test_bin[:, label_dict[i]],
        y_score[:, label_dict[i]]
    )
    auc_i = auc(fpr_i, tpr_i)
    axes[1].plot(fpr_i, tpr_i, lw=1, label=f'Class {i} (AUC = {auc_i:.7f})')

axes[1].plot([0, 1],[0, 1], "k--", label="Random guess (AUC = 0.5)")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].legend(loc="Best")
axes[1].set_title("ROC curve (OvR)")
plt.savefig("PR_ROC_RF.svg")
plt.show()
```

### 2.8. Завантаження моделі RF

### 2.9. Створення та візуалізація матриці помилок для моделі класифікатора випадкового лісу

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay

np.set_printoptions(precision=2)
plt.rcParams["figure.figsize"] = (14,7)

title = "Random Forest Classifier Confusion Matrix"
disp = ConfusionMatrixDisplay.from_estimator(
    rand_forest,
    X_test,
    y_test,
    cmap=plt.cm.Blues,
    normalize="true",
)
disp.ax_.set_title(title)
plt.savefig("RandomForest-Confusion-Matrix.svg")
plt.show()
```

### 3.1. Тренування моделі класифікатора опорних векторів (SVC) з радіальною базовою функцією (RBF) на тренувальному наборі даних

```python
from sklearn.svm import SVC

svc = SVC(kernel="rbf")
svc.fit(X_train, y_train)
train_score = svc.score(X_train, y_train)
test_score = svc.score(X_test, y_test)

print(f'Accuracy on training set: {train_score:.2%}')
print(f'Accuracy on test set: {test_score:.2%}')
```

### 3.2. Виконання прогнозування на тестовому наборі даних. Друк звіту про класифікацію

```python
y_pred = svc.predict(X_test)
print(classification_report(y_test, y_pred, digits=4))
```

### 3.3. Використання GridSearchCV для оптимізації гіперпараметрів C та gamma моделі SVM з радіальною базовою функцією (rbf). Тренування її на тренувальному наборі даних.

```python
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1e-4, 1e-3, 1e-2]
}

CV_svc = GridSearchCV(SVC(kernel='rbf'), param_grid, n_jobs=-1)
CV_svc.fit(X_train, y_train)
```

### 3.4. Отримані найкращі гіперпараметри для моделі SVM

```python
CV_svc.best_params_
```

### 3.5. Тренування оптимізаваної моделі SVM за допомогою GridSearchCV на тренувальному наборі даних та оцінка її точність на тренувальному та тестовому наборах

```python
y_pred = CV_svc.predict(X_test)
print(classification_report(y_test, y_pred, digits=4))
```

### 3.6. Тренування моделі SVM з оптимальними параметрами для класифікації бінаризованих міток, візуалізація криві точності-повноти та ROC для кожного класу, і зберігання графіків у файл SVG

```python
y_train_bin = label_binarize(y_train, classes=labels)
y_test_bin = label_binarize(y_test, classes=labels)

clf = OneVsRestClassifier(SVC(kernel='rbf', **CV_svc.best_params_))

clf.fit(X_train, y_train_bin)
y_score = clf.decision_function(X_test)
fig, axes = plt.subplots(
    nrows=1,
    ncols=2,
    figsize=(12, 6),
    layout="constrained",
)


# precision recall curve
ap = average_precision_score(y_test_bin, y_score, average=None)
for i in labels:
    precision_i, recall_i, _ = precision_recall_curve(
        y_test_bin[:, label_dict[i]],
        y_score[:, label_dict[i]]
    )
    axes[0].plot(recall_i, precision_i, lw=2, label=f'Class {i} (AP = {ap[label_dict[i]]:.7f})')
    
axes[0].set_xlabel("Recall")
axes[0].set_ylabel("Precision")
axes[0].legend(loc="Best")
axes[0].set_title("Precision vs Recall curve (OvR)")

for i in labels:
    fpr_i, tpr_i, _ = roc_curve(
        y_test_bin[:, label_dict[i]],
        y_score[:, label_dict[i]]
    )
    auc_i = auc(fpr_i, tpr_i)
    axes[1].plot(fpr_i, tpr_i, lw=1, label=f'Class {i} (AUC = {auc_i:.7f})')

axes[1].plot([0, 1],[0, 1], "k--", label="Random Guess (AUC = 0.5)")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].legend(loc="Best")
axes[1].set_title("ROC Curve (OvR)")
plt.savefig("PR_ROC_SVC.svg")
plt.show()
```

# Decision Tree

### 4.1. Тренування моделі класифікатора Decision Tree на тренувальному наборі даних

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Initialize the Decision Tree classifier
dt = DecisionTreeClassifier()

# Fit the model to the training data
dt.fit(X_train, y_train)

# Calculate the accuracy on the training set and test set
train_accuracy = dt.score(X_train, y_train)
test_accuracy = dt.score(X_test, y_test)

print(f'Accuracy on training set: {train_accuracy:.2%}')
print(f'Accuracy on test set: {test_accuracy:.2%}')
```

### 4.2. Виконання прогнозування на тестовому наборі даних. Друк звіту про класифікацію

```python
# Predict on the training set and the test set
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)

# Generate and print the classification report for the test set
print(classification_report(y_test, y_test_pred, digits=4))
```

### 4.3. Використання GridSearchCV для оптимізації гіперпараметрів моделі Decision Tree. Тренування її на тренувальному наборі даних. Отримані найкращі гіперпараметри

```python
# Parameter tuning with GridSearchCV for Decision Tree
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

CV_dt = GridSearchCV(DecisionTreeClassifier(), param_grid_dt, n_jobs=-1)
CV_dt.fit(X_train, y_train)

# Best parameters and scores
print("Best parameters found: ", CV_dt.best_params_)
print("Best accuracy found: ", CV_dt.best_score_)
```

### 3.5. Тренування оптимізаваної моделі Decision Tree за допомогою GridSearchCV на тренувальному наборі даних та оцінка її точність на тренувальному та тестовому наборах

```python
y_pred_dt = CV_dt.predict(X_test)

print(classification_report(y_test, y_pred_dt, digits=4))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import average_precision_score, precision_recall_curve, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
import matplotlib.pyplot as plt

clf_dt = OneVsRestClassifier(DecisionTreeClassifier(max_depth=CV_dt.best_params_['max_depth'],
                                                    criterion=CV_dt.best_params_['criterion'],
                                                    min_samples_split=CV_dt.best_params_['min_samples_split'],
                                                    min_samples_leaf=CV_dt.best_params_['min_samples_leaf']))
clf_dt.fit(X_train, y_train_bin)

y_score_dt = clf_dt.predict_proba(X_test)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6), layout="constrained")

ap_dt = average_precision_score(y_test_bin, y_score_dt, average=None)
for i in labels:
    precision_i, recall_i, _ = precision_recall_curve(y_test_bin[:, label_dict[i]], y_score_dt[:, label_dict[i]])
    axes[0].plot(recall_i, precision_i, lw=2, label=f'Class {i} (AP = {ap_dt[label_dict[i]]:.7f})')
    
axes[0].set_xlabel("Recall")
axes[0].set_ylabel("Precision")
axes[0].legend(loc="best")
axes[0].set_title("Precision vs Recall curve (OvR)")

for i in labels:
    fpr_i, tpr_i, _ = roc_curve(y_test_bin[:, label_dict[i]], y_score_dt[:, label_dict[i]])
    auc_i = auc(fpr_i, tpr_i)
    axes[1].plot(fpr_i, tpr_i, lw=1, label=f'Class {i} (AUC = {auc_i:.7f})')

axes[1].plot([0, 1], [0, 1], "k--", label="Random Guess (AUC = 0.5)")
axes[1].set_xlabel("False Positive Rate")
axes[1].set_ylabel("True Positive Rate")
axes[1].legend(loc="best")
axes[1].set_title("ROC Curve (OvR)")

plt.savefig("PR_ROC_DT.svg")
plt.show()
```

### Результати аналізу

У дослідженнях були навчені та оптимізовані 3 моделі: Random Forest, SVC з ядром RBF, Decision Tree. Оптимізація відбувалася за рахунок використання методу перехресної перевірці пошуку у сітці(Grid search cross-validation).  Цей метод дозволяє навчати моделі з діапазоном різних значень гіперпараметрів, з оцінкою продуктивності моделі за допомогою перехресної перевірки та з вибором набору гіперпараметрів, які є найоптимальнішими.
Наприклад, якщо сітка для моделі включає гіперпараметри C і γ, і C має можливі значення [0,1, 1, 10], а γ має можливі значення [0,01, 0,1, 1], тоді перехресна перевірка пошуку в сітці навчатиметься та оцінити модель, використовуючи всі можливі комбінації цих значень, в результаті чого в цілому виходить 9 різних моделей. 
- Random Forest показує найкращі результати за всіма критеріями, включаючи точність, середню точність, відклик, та F1-оцінку. Це робить її найкращим варіантом для даного набору даних з найвищою загальною точністю 99.98%. Найкращі обрані гіперпараметри для цієї моделі: criterion=entropy, max_features = 0.5,  n_estimators = 200.
- SVС показує найнижчі результати серед трьох моделей, з точністю 97.49% та найнижчими середніми значеннями за іншими метриками. Незважаючи на те, що SVС часто ефективний для різних задач класифікації, у цьому конкретному випадку він відстає від інших моделей. Найкращі обрані гіперпараметри для цієї моделі: C = 1000, gamma = 0.01.
- Decision Tree демонструє результати, дуже близькі до Random Forest, з невеликим зниженням точності до 99.97%. Це підтверджує, що для деяких наборів даних простіші моделі можуть наблизитись до або навіть перевершити складніші ансамблі моделей за певних обставин. Найкращі обрані гіперпараметри для цієї моделі: criterion = gini, max_depth = 30, min_samples_leaf = 1, min_samples_split = 2.
